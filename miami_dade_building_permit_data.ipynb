### This cell creates a new DF

```python
df2 = df.filter(['ApplicationNumber','CompanyName','DaysInCityNumeric','FirstSubmissionDate','FolioNumber','IsPermitFinal','PlanAcceptedDate','PlanCreatedDate','PropertyType','TotalDaysInPlanReviewNumeric'])
```

### Create a column for the number of times a company has applied for a permit since 2014

```python
df2['number_of_permits_applied_for'] = df2.groupby(['CompanyName'])['TotalDaysInPlanReviewNumeric'].transform('count')
```

```python
### This line of code checks to make sure the 'Company_Counts' value is the same for all companies
#df2.loc[df2['CompanyName'] == 'Miller Development Group']
```

```python
df2.CompanyName.value_counts()
```

```python
### Cleans commas from the column
df2['TotalDaysInPlanReviewNumeric'] = df2['TotalDaysInPlanReviewNumeric'].replace(',','',regex=True)

### Converts the column to numeric
df2.TotalDaysInPlanReviewNumeric = pd.to_numeric(df2.TotalDaysInPlanReviewNumeric)

### Checks the data types
df2.dtypes
```

```python
### Calculates the mean based on company name
df2['PlanReviewAverageDays'] = df2.groupby('CompanyName',as_index=False)['TotalDaysInPlanReviewNumeric'].transform('mean')
```

```python
df2.loc[df2['CompanyName'] == 'THYSSENKRUPP ELEVATOR CORP  MIAMI']
```

```python
df3 = df2.filter(['CompanyName','number_of_permits_applied_for','PlanReviewAverageDays'])
print(len(df3))
```

```python
df3 = df3.drop_duplicates()
print(len(df3))
```

```python
#pd.set_option('display_max_rows',500)
```

```python
df3 = df3.sort_values(by='number_of_permits_applied_for',ascending=False)
df3 = df3.reset_index(drop=True)
df3
```

```python
df3.to_csv('preliminary_permit_data.csv')
```
